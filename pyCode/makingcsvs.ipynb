## inputs:

- the output of the R code that has the unchanged sequence names/regions and repeat weights

### - PF0023_2reps_correctednogaps 
- the output of the R code that has the 64 residue sequences and unchanged sequence names/regions

## outputs:
### - weights_newnames.txt 
- has plm.c readable sequence names and repeat weights

### - numbers.txt 
- just the repeat weights without the readable sequence names. This will be the input to the C code
- I confirmed that it does not matter to use this method because the order is the same in the weights_newnames.txt and the input msa to plm.c

### - PF0023_2reps_input_plm.txt 
- the input MSA to the plm.c, has 64 residues and 91143 sequences

```python
# importing packages
import re
import pandas as pd
import numpy as np
```

```python
# create dictionary of original names and new names
# save out new csv with new name
df_repweights = pd.read_csv('~/Dropbox (Harvard University)/Debbie-Ada/repeatsProject/rCode/weights_rownamestrue.txt',header=None,delimiter='\s',names=['names','weights'])
df_repweights['names']=df_repweights['names'].str.replace('"','')
df_repweights['sequence_id'] = df_repweights['names'].str.split(r"/[\d']",n=1,expand=True)[0]
tmp = 'whatever'
tmp2 = np.zeros_like(df_repweights.weights.values,dtype=int)
num = 0
for (index,row) in df_repweights.iterrows():
        if row.sequence_id != tmp:
            num = 0
            tmp = row.sequence_id
        num += 1
        tmp2[index] = num
df_repweights['repeat_num'] = tmp2
df_repweights['new_names'] = df_repweights['sequence_id'].map(str) +'_'+ df_repweights['repeat_num'].map(str)+'/1-64'
name_dict=dict(zip(df_repweights.names.values,df_repweights.new_names.values))
df_repweights[['new_names','weights']].to_csv('/Users/adashaw/Dropbox (Harvard University)/Debbie-Ada/repeatsProject/pyCode/new_names_weights.txt',header=None,index=False)
```

```python
# I gave up on the mixed data csv - too fucking difficult to implement in C
# instead I'll just send the weights into the plm.c code and then do the checking outside

df_repweights[['weights']].to_csv('numbers.txt',header=None,index=False)
```

```python
# read in the msa and rewrite the names

msa_file = '/Users/adashaw/Dropbox (Harvard University)/Debbie-Ada/repeatsProject/rCode/PF0023_2reps_correctednogaps.txt'
with open(msa_file,'r') as file:
    all_lines=file.readlines()
for (index,lines) in enumerate(all_lines):
    if lines.startswith('>'):
        all_lines[index] = all_lines[index].replace(all_lines[index][1:-1],name_dict[all_lines[index][1:-1]])
with open('/Users/adashaw/Dropbox (Harvard University)/Debbie-Ada/repeatsProject/pyCode/PF0023_2reps_plm_input.txt','w') as file:
    file.writelines(all_lines)
```
