```python
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "PF00023_ncbi.txt"
def readAlignment(file):
    with open(file) as f:
        lines = f.readlines()
    line_numbers = [i for i in range(len(lines)) if '>' in lines[i] ]

    names = [lines[i].replace('\n','').replace('>','') for i in line_numbers]
    sequences = [''.join(lines[i+1:i+4]).replace('\n','') for i in line_numbers]
    msa = np.zeros([len(names),len(sequences[0])+1],dtype=np.object_)   
    msa[:,0] = names
    for tick_l in range(len(sequences)):
        msa[tick_l,1:] = list(sequences[tick_l])
    return msa
```

```python
def msa_remove_gaps(msa,cutoff_gaps=0.7):
    gaps = np.sum(msa=='-',axis=0)/(len(msa))
    return msa[:,gaps<cutoff_gaps]
```

```python
def names_split(msa):
    names = msa[:,0]
    names_df = pd.DataFrame(names,)
    names_df=names_df[0].str.split('/',n=1,expand=True)
    names_df[[1,2]]=names_df[1].str.split('-',n=1,expand=True)
    return names_df
def MSA_n_neighbors(msa,n=1,outfile=None):
    names=names_split(msa)
    f=pd.DataFrame(names[0].value_counts())
    f[f[0]>n]
```
