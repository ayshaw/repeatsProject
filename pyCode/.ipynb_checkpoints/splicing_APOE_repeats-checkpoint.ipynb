```python
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
```

```python
def readAlignment(file):
    with open(file) as f:
        lines = f.readlines()
    line_numbers = [i for i in range(len(lines)) if '>' in lines[i] ]
    names = [lines[i].replace('\n','').replace('>','') for i in line_numbers]
    sequences = [''.join(lines[i+1:i+5]).replace('\n','') for i in line_numbers]
    msa = np.zeros([len(names),len(sequences[0])+1],dtype=np.object_)   
    msa[:,0] = names
    for tick_l in range(len(sequences)):
        msa[tick_l,1:] = list(sequences[tick_l])
    return msa
```

```python
msa = readAlignment('/home/as974/ada/repeatsProject/APOE_HUMAN_full_rerun_b025/APOE_HUMAN_full_rerun_b025.a2m')
```

```python
msa_df = pd.DataFrame(msa,columns = ['name']+list(map(str,np.arange(1,318)))) 
msa_df.head()
```
